```python
def clean_text(text):
    if isinstance(text, str):
        text = text.lower()
        tokens = [i for i in nltk.word_tokenize(text) if i.isalnum()]
        stop_words = set(stopwords.words('indonesian') + list(string.punctuation))
        tokens = [i for i in tokens if i not in stop_words]

        return " ".join(tokens)
    else:
        return str(text)

df['Cleaned_Text'] = df['Teks'].apply(clean_text)
df.head()
```

```python
# Daftar kata-kata yang ingin dihapus
kata_kata_tidak_penting = ['eh', 'ee', 'terima', 'kasih', 'ya','mesti']

# Fungsi untuk menghapus kata-kata tidak penting
def remove_common_words(text, common_words):
    words = text.split()
    cleaned_words = [word for word in words if word not in common_words]
    return ' '.join(cleaned_words)

# Menghapus kata-kata tidak penting dari setiap teks
df['Cleaned_Text'] = df['Cleaned_Text'].apply(lambda x: remove_common_words(x, kata_kata_tidak_penting))
```

```python
# Identifikasi kata-kata kunci yang paling sering digunakan oleh setiap pembicara dalam setiap topik
def top_words_per_speaker(dataframe, pembicara):
    # Menggabungkan semua teks dari pembicara tertentu
    text = ' '.join(dataframe[dataframe['Pembicara'] == pembicara]['Cleaned_Text'])
    # Memecah teks menjadi kata-kata
    words = text.split()
    # Menghitung frekuensi kemunculan setiap kata
    word_freq = pd.Series(words).value_counts()
    # Mengembalikan 10 kata paling umum
    return word_freq.head(10)

# Contoh pemanggilan fungsi untuk masing-masing pembicara
print("Top 10 Kata-kata untuk Anis:")
print(top_words_per_speaker(df, 'Anis'))
print("Top 10 Kata-kata untuk Prabowo:")
print(top_words_per_speaker(df, 'Prabowo'))
print("\nTop 10 Kata-kata untuk Ganjar:")
print(top_words_per_speaker(df, 'Ganjar'))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt

# Inisialisasi TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fungsi untuk mendapatkan top N kata-kata yang paling penting untuk seorang pembicara
def get_top_words_tfidf(dataframe, tfidf_vectorizer, pembicara, n=10):
    # Menghitung TF-IDF hanya untuk teks dari pembicara yang sesuai
    tfidf_matrix = tfidf_vectorizer.fit_transform(dataframe[dataframe['Pembicara'] == pembicara]['Cleaned_Text'])
    # Mendapatkan daftar fitur baru
    feature_names = tfidf_vectorizer.get_feature_names_out()
    # Menghitung nilai rata-rata TF-IDF untuk setiap kata
    avg_tfidf = tfidf_matrix.mean(axis=0).A1
    # Mengurutkan indeks kata berdasarkan nilai rata-rata TF-IDF
    sorted_indices = avg_tfidf.argsort()[::-1]
    # Mengambil top N kata-kata
    top_words = [feature_names[idx] for idx in sorted_indices[:n]]
    # Mengambil skor TF-IDF untuk setiap kata
    top_scores = [avg_tfidf[idx] for idx in sorted_indices[:n]]
    return top_words, top_scores

# Daftar pembicara yang unik
pembicara_list = df['Pembicara'].unique()

# Membuat plot top 10 kata-kata yang paling penting untuk setiap pembicara
for pembicara in pembicara_list:
    top_words, top_scores = get_top_words_tfidf(df, tfidf_vectorizer, pembicara)
    plt.figure(figsize=(10, 6))
    plt.barh(top_words, top_scores, color='skyblue')
    plt.title('Top 10 Kata-kata Penting untuk ' + pembicara)
    plt.xlabel('TF-IDF Score')
    plt.ylabel('Kata')
    plt.gca().invert_yaxis()
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```
